```python
# removing units

df2["Mileage"]=df2["Mileage"].str.replace('km/kg','')
df2["Mileage"]=df2["Mileage"].str.replace('kmpl','')
df2["Engine"]=df2["Engine"].str.replace('CC','')
df2["Power"]=df2["Power"].str.replace('bhp','')
df2
```

```python
# null =====> 0 (replacing null values from mileage ,engine,power)
df2["Mileage"]=df2["Mileage"].str.replace('null','0')
df2["Power"]=df2["Power"].str.replace('null','0')
df2["Engine"]=df2["Engine"].str.replace('null','0')
df2
```

```python
df2.dtypes
```

```python
# converting datatypes (power,engine,mileage)

df2["Mileage"]=df2["Mileage"].astype(float)
df2["Engine"]=df2["Engine"].astype(float) # int also applicable
df2["Power"]=df2["Power"].astype(float)
df2
```

```python
df2.dtypes
```

```python
# power, mileage,engine can't have zero as its values, need to convert to missing value

# how to convert numerical data to missing values

df2.loc[df2.Mileage==0,'Mileage']=np.nan  #nan==missing values
df2.loc[df2.Power==0,'Power']=np.nan
df2.loc[df2.Engine==0,'Engine']=np.nan
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
x_test
```

```python
y_test
```

```python
x_train
```

```python
y_train
```

```python
plt.scatter(x,y)
plt.xlabel("Height of Father",size=20)
plt.ylabel("Height of Son",size=20)
plt.title("Father - Son height",size=30)
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
plt.scatter(x_train,y_train,color='r')
plt.plot(x_test,y_pred)
plt.xlabel("Father height")
plt.ylabel("Son height")
plt.title("Father-Son height",size=30)
```

```python
print(model.predict([[70]]))
```

```python
print("Slope =",model.coef_)
```

```python
print("Constant=",model.intercept_)
```

```python
df1=pd.DataFrame({"Actual value":y_test,"predicted value":y_pred,"difference":y_test-y_pred})
df1
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,root_mean_squared_error,r2_score
mae=mean_absolute_error(y_test,y_pred)
mape=mean_absolute_percentage_error(y_test,y_pred)
mse=mean_absolute_error(y_test,y_pred)
rmse=root_mean_squared_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
mae
```

```python
import numpy as np
import pandas as pd
df=pd.read_csv("customer1.txt",header=None)
df.columns=["Id","fname","lname","age","prof","loc"]
df
```

```python
a=df["loc"].value_counts()
a
```

```python
import matplotlib.pyplot as plt
plt.bar(a.index,a.values)  # index ===> x axis , values ===> y axis
plt.xlabel("Location")
plt.ylabel("Count")
plt.title("Country Counts",fontsize=20,color='')
```

```python
plt.pie(a.values,labels=a.index,autopct="%1.0f%%")
plt.legend(loc="upper right")
plt.tight_layout()
plt.title("Location Count",fontsize=20)
```

```python
x=[1,5,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y,'*') # * is given gor getting points only
plt.show()
```

```python
x=[1,5,1,4,5]
y=[6,7,8,2,10]
plt.plot(x,y,marker='o')   # it will give points also
plt.show()
```

```python
plt.plot(x,y,'*:g')   # * as marker , : as line type , r as red color  (diff lines - solid line, dotted line , dashed line(--))
plt.show()
```

```python
plt.plot(x,y,'o--c')   # * as marker , : as line type(dotted) , r as red color(k for black)  (diff lines - solid line, dotted line , dashed line(--))
plt.show()
```

```python
plt.plot(x,y,marker='o',ms=50)  # ms ===> marker size
plt.show()
```

```python
plt.plot(x,y,marker='o',ms=20,mfc='r')  # mfc ===> marker face color
plt.show()
```

```python
plt.plot(x,y,marker='o',ms=20,mec='r')  # mec ===> marker edge color
plt.show()
```

```python
# Line properties

plt.plot(x,y,marker='o',linestyle='dashed',linewidth=4,color='g')
```

```python
plt.plot(x,y,marker='o',ms=15,mfc='k',mec='r',linestyle='dashed',linewidth=4,color='g')
```

```python
height=[145,147,146,150,155,153,158]
weight=[60,56,45,50,60,48,53]
plt.plot(height,weight)
plt.xlabel("Height in cm",color='r',fontsize=10)
plt.ylabel("Weight in kg",color='b',fontsize=10)
plt.title("Height - Weight Graph",color='g',fontsize=15)
```

```python
velocity=[145,150,146,156,134,155,152]
displacement=[30,47,78,56,34,55,56]
plt.plot(velocity,displacement)
plt.xlabel("Velocity in m/s",color='r',fontsize=10)
plt.ylabel("displacement in km",color='g',fontsize=10)
plt.title("Velocity - Displacement Graph",color='b',fontsize=20)
plt.grid(axis='x',linestyle='dotted',linewidth=3,color='c')   # grid() is a line like in graph paper
# grid(axis='x') --- only the lines corresponding to the x axis will show
# grid(axis='y') --- only the lines corresponding to the y axis will show
```

```python
# scatter plot -- to represent points in a graph (like marker) , scatterplot is used
x1=[1,3,5,7,9]
y1=[2,3,6,4,8]
x2=[4,3,9,1,8]
y2=[2,7,3,9,4]
plt.scatter(x1,y1,color='r')
plt.scatter(x2,y2,color='c')
```

```python
# sub plot --- dividing the graph into two parts and first part will use for graph1 and second part will used for graph2
x1=[1,3,5,7,9]
y1=[2,3,6,4,8]
plt.subplot(1,2,1) # --- subplot(row,column,position)
plt.plot(x1,y1)
plt.xlabel("x_axis",color='b',fontsize=10)
plt.ylabel("y_axis",color='b',fontsize=10)
plt.title("Graph 1")

x2=[4,3,9,1,8]
y2=[2,7,3,9,4]
plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.ylabel("y_axis",color='b',fontsize=10)
plt.ylabel("y_axis",color='b',fontsize=10)
plt.title("Graph 2")
plt.supertitle("Graph1-Graph2") # supertitle is the main title
```

```python
# bar chart

x1=[1,3,5,7,9]
y1=[2,3,6,4,8]
plt.subplot(1,2,1)
plt.bar(x1,y1,color='r')

x2=['India','China','America','Russia','Australia']
y2=[2,3,6,4,8]
plt.subplot(1,2,2)
plt.bar(x2,y2,color='c')
# barh ----for horizontal
```

```python
# histogram  --- no. of frequency observation (only one axis)
x=[1,1,2,2,2,4,4,4,4,5,5,5,6,6,6,6,6,3,3,3,3,7,7,7,7,7,7]
plt.hist(x)
```

```python
x1=[10,20,30,40]
lab=['Apple','Orange','Mango','Grapes'] # for labels
colo=['red','yellow','green','cyan']
plt.pie(x1,labels=lab,colors=colo,autopct='%1.0f%%')
plt.legend(loc='lower right') # for specific color along with label
plt.tight_layout()  # to avoid overlapping
```
